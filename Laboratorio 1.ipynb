### Problema 3

Cree una función que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su función debe crear 3 imágenes como salida. Para entregar este ejercicio debe incluir una las imágenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema #2.

```python
def create_grayscale_3d(image_path):
    # Cargar la imagen original
    original_image = Image.open(image_path)
    
    # Convertir la imagen a escala de grises
    grayscale_image = original_image.convert('L')
    
    # Obtener la matriz de píxeles de la imagen en escala de grises
    grayscale_array = np.array(grayscale_image)
    
    # Crear una imagen en tres dimensiones duplicando el array de escala de grises
    grayscale_3d_array = np.stack((grayscale_array,) * 3, axis=-1)
    
    # Convertir la matriz de tres dimensiones en una imagen
    grayscale_3d_image = Image.fromarray(grayscale_3d_array)
    
    # Guardar cada canal como una imagen individual
    channel_r, channel_g, channel_b = grayscale_3d_image.split()
    channel_r.save("grayscale_channel_r.jpg")
    channel_g.save("grayscale_channel_g.jpg")
    channel_b.save("grayscale_channel_b.jpg")
```

```python
create_grayscale_3d("LBK.jpg")
```

```python
def merge_grayscale_channels(channel_r_path, channel_g_path, channel_b_path):
    channel_r = Image.open(channel_r_path)
    channel_g = Image.open(channel_g_path)
    channel_b = Image.open(channel_b_path)
    
    return Image.merge("RGB", (channel_r, channel_g, channel_b))


merged_grayscale_image = merge_grayscale_channels("grayscale_channel_r.jpg", "grayscale_channel_g.jpg", "grayscale_channel_b.jpg")
```
